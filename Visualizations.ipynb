```python
houseprices = pd.read_csv("houseprices.csv")
houseprices.info()
```

```python
#Illustrate the distribution of saleprice.
saleprice_hist = sns.distplot(houseprices['saleprice'], bins = 20)
plt.xticks([0,100000, 200000, 300000, 400000, 500000, 600000, 700000, 800000], 
           ['0','100k', '200k', '300k', '400k', '500k', '600k', '700k', '800k'])
plt.xlabel('SALEPRICE')
```

```python
#Illustrate the distribution of the lotarea.
lotarea_hist = sns.distplot(houseprices['lotarea'])
plt.xlabel('LOTAREA')
```

```python
#Use a boxplot to illustrate how overallcond affects the sale price.
houseprices.centralair
```

```python
#Use a boxplot to illustrate how overallcond affects the sale price.
sns.boxplot(y = "saleprice",x = "overallcond", data = houseprices)
#it looks like the saleprice went up when overallcond went up
```

```python
#On the same boxplot, highlight the difference between having central air conditioning or not.
sns.boxplot(y = "saleprice", hue = "centralair",x = "overallcond", data = houseprices)
# it appears that sale price is higher with central aircondition
```

```python
#Is there a correlation between lotarea and the saleprice?
sns.scatterplot(x = "lotarea", y = "saleprice", data = houseprices)
#it appears that there is no correlation between lotarea and saleprice
```
